```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn import datasets
```

```python
A, b = datasets.make_classification(n_samples=75, n_features=7, n_classes=3, n_redundant=0,n_clusters_per_class=1, weights=[0.5, 0.3, 0.2])
A = pd.DataFrame(A);
A['Country'] = ["China","India","Japan","Spain","Italy"]*15;
A['Class'] = b
A.head(10)
```

```python
print (A.dtypes)
```

```python
A.describe()
```

```python
A.rank()
```

```python
A.corr()
```

```python
import matplotlib.pyplot as plt
A['Class'].hist(bins=10)
A.boxplot(column='Class', by = 'Country')
x = A["Class"]
y = A["Country"]
plt.scatter(x, y, label= "stars", color= "m", 
            marker= "*", s=30)
plt.xlabel('Class')
plt.ylabel('Country')
plt.show()
```

```python
plt.scatter(A['Class'], A["Country"])
plt.title("Scatter Plot")
plt.xlabel('Class')
plt.ylabel('Country')
  
plt.show()
```

```python
plt.plot(A['Class'])
plt.plot(A['Country'])
  
# Adding Title to the Plot
plt.title("Scatter Plot")
  
# Setting the X and Y labels
plt.xlabel('Class')
plt.ylabel('Country')
  
plt.show()
```

```python
plt.bar(A['Class'], A["Country"])
plt.title("Scatter Plot")
plt.xlabel('Class')
plt.ylabel('Country')
  
plt.show()
```

```python
corr = A.corr()
corr
```

```python
sns.heatmap(corr)
```

```python
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(7, 5))
    ax = sns.heatmap(corr, mask=mask, vmax=1, square=True)
```

```python
g = sns.PairGrid(corr)
g.map(plt.scatter);
```
